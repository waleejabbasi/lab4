```python
# Lab task 1
#2022f-bse-123
import pandas as pd

# Step 1: Create the dataset
data = {
    'Weather': ['Sunny', 'Sunny', 'Overcast', 'Rainy', 'Rainy', 'Rainy', 'Overcast', 'Sunny', 'Sunny', 'Rainy', 'Sunny', 'Overcast', 'Overcast', 'Rainy'],
    'Temperature': ['Hot', 'Hot', 'Hot', 'Mild', 'Cool', 'Cool', 'Cool', 'Mild', 'Cool', 'Mild', 'Mild', 'Mild', 'Hot', 'Mild'],
    'Play': ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'Yes', 'Yes', 'Yes', 'No']
}

# Convert the data into a pandas DataFrame
df = pd.DataFrame(data)

# Step 2: Calculate prior probabilities for Play = Yes and Play = No
P_Play_Yes = len(df[df['Play'] == 'Yes']) / len(df)
P_Play_No = len(df[df['Play'] == 'No']) / len(df)

# Step 3: Calculate conditional probabilities for Weather and Temperature given Play
# P(Weather = Overcast | Play = Yes)
P_Weather_Overcast_Play_Yes = len(df[(df['Weather'] == 'Overcast') & (df['Play'] == 'Yes')]) / len(df[df['Play'] == 'Yes'])
# P(Temperature = Mild | Play = Yes)
P_Temperature_Mild_Play_Yes = len(df[(df['Temperature'] == 'Mild') & (df['Play'] == 'Yes')]) / len(df[df['Play'] == 'Yes'])

# P(Weather = Overcast | Play = No)
P_Weather_Overcast_Play_No = len(df[(df['Weather'] == 'Overcast') & (df['Play'] == 'No')]) / len(df[df['Play'] == 'No'])
# P(Temperature = Mild | Play = No)
P_Temperature_Mild_Play_No = len(df[(df['Temperature'] == 'Mild') & (df['Play'] == 'No')]) / len(df[df['Play'] == 'No'])

# Step 4: Calculate the likelihood for Play = Yes and Play = No
P_Play_Yes_given_Weather_Overcast_and_Temperature_Mild = P_Weather_Overcast_Play_Yes * P_Temperature_Mild_Play_Yes * P_Play_Yes
P_Play_No_given_Weather_Overcast_and_Temperature_Mild = P_Weather_Overcast_Play_No * P_Temperature_Mild_Play_No * P_Play_No

# Step 5: Normalize to calculate the final probabilities
total = P_Play_Yes_given_Weather_Overcast_and_Temperature_Mild + P_Play_No_given_Weather_Overcast_and_Temperature_Mild

# Probability for Play = Yes
P_Play_Yes_normalized = P_Play_Yes_given_Weather_Overcast_and_Temperature_Mild / total
# Probability for Play = No
P_Play_No_normalized = P_Play_No_given_Weather_Overcast_and_Temperature_Mild / total

# Step 6: Make the prediction
prediction = 'Yes' if P_Play_Yes_normalized > P_Play_No_normalized else 'No'

# Output the result
print(f"P(Play = Yes | Weather = Overcast, Temperature = Mild) = {P_Play_Yes_normalized}")
print(f"P(Play = No | Weather = Overcast, Temperature = Mild) = {P_Play_No_normalized}")
print(f"Prediction: Can the player play? {prediction}")

```

```python
#task2
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.preprocessing import LabelEncoder

# Step 1: Prepare the dataset
data = {
    'age': ['youth', 'youth', 'middle aged', 'senior', 'senior', 'senior', 'senior', 'middle-aged', 'youth', 'youth', 'youth', 'youth', 'youth', 'youth', 'senior', 'senior', 'middle aged', 'middle-aged', 'senior'],
    'income': ['high', 'high', 'high', 'medium', 'low', 'low', 'low', 'low', 'medium', 'low', 'medium', 'low', 'medium', 'high', 'medium', 'medium', 'medium', 'medium', 'high'],
    'student': ['no', 'no', 'no', 'no', 'yes', 'yes', 'yes', 'yes', 'no', 'yes', 'yes', 'yes', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'yes'],
    'credit_rating': ['fair', 'excellent', 'fair', 'fair', 'fair', 'fair', 'excellent', 'fair', 'fair', 'fair', 'excellent', 'fair', 'excellent', 'fair', 'fair', 'fair', 'excellent', 'excellent', 'fair'],
    'buys_computer': ['no', 'no', 'yes', 'yes', 'yes', 'yes', 'no', 'yes', 'no', 'yes', 'no', 'yes', 'yes', 'yes', 'yes', 'no', 'yes', 'no', 'no']
}

# Convert the data into a pandas DataFrame
df = pd.DataFrame(data)

# Step 2: Preprocess the data by encoding categorical variables
label_encoder = LabelEncoder()
df['age'] = label_encoder.fit_transform(df['age'])
df['income'] = label_encoder.fit_transform(df['income'])
df['student'] = label_encoder.fit_transform(df['student'])
df['credit_rating'] = label_encoder.fit_transform(df['credit_rating'])
df['buys_computer'] = label_encoder.fit_transform(df['buys_computer'])

# Step 3: Split the data into training and testing sets
X = df.drop('buys_computer', axis=1)  # Features
y = df['buys_computer']  # Target variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Train the Naive Bayes classifier
model = GaussianNB()
model.fit(X_train, y_train)

# Step 5: Make predictions
y_pred = model.predict(X_test)

# Step 6: Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

# Step 7: Output the results
print("Accuracy Score:", accuracy)
print("Confusion Matrix:")
print(conf_matrix)

# Interpretation of the confusion matrix
TP = conf_matrix[1, 1]  # True Positive
TN = conf_matrix[0, 0]  # True Negative
FP = conf_matrix[0, 1]  # False Positive
FN = conf_matrix[1, 0]  # False Negative

print("\nConfusion Matrix Interpretation:")
print(f"True Positives (TP): {TP}")
print(f"True Negatives (TN): {TN}")
print(f"False Positives (FP): {FP}")
print(f"False Negatives (FN): {FN}")

# Calculate the accuracy manually
calculated_accuracy = (TP + TN) / (TP + TN + FP + FN)
print(f"Calculated Accuracy: {calculated_accuracy}")
```

```python
# Home task
#2022f-bse-123

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Creating the dataset
data = {
    'gender': [0, 1, 1, 1, 0, 1],  # male=0, female=1
    'insurance': [1, 0, 1, 1, 0, 1],  # Yes=1, No=0
    'number_of_visits': [2, 1, 5, 3, 0, 7],
    'emergency_room_visit': [0, 1, 1, 0, 0, 1],  # Yes=1, No=0
    'major_disease': [0, 1, 2, 3, 4, 5],  # Encoding diseases (Flu=0, Asthma=1, etc.)
    'has_chronic_condition': [0, 1, 1, 1, 0, 1]  # Yes=1, No=0
}

df = pd.DataFrame(data)

# Features and target variable
X = df[['gender', 'insurance', 'number_of_visits', 'emergency_room_visit', 'major_disease']]
y = df['has_chronic_condition']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# Initialize the Na√Øve Bayes classifier
model = GaussianNB()

# Train the model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Example prediction for a new patient
# Gender: Female (1), Insurance: Yes (1), Number of Visits: 4, ER Visit: Yes (1), Major Disease: Hypertension (3)
new_patient = [[1, 1, 4, 1, 3]]
prediction = model.predict(new_patient)
print(f"Prediction for the new patient: {'Has chronic condition' if prediction[0] == 1 else 'No chronic condition'}")
```
